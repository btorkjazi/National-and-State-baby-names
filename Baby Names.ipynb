The first thing that I'd like to analyze is to find out about the dominant gender in all over these years. So, I develop following code to find the total number of each gender in the USA and then plot a pie chart representing this result in an easier way.

Next step is to find the dominant gender in each year.

```python
#Number of boys vs girls in each year
gender_year = national_names.groupby(["Year", "Gender"])["Count"].sum()
gender_year1 = gender_year.reset_index()
gender_year1
```

```python
boys = gender_year1[gender_year1["Gender"] == "M"]
girls = gender_year1[gender_year1["Gender"] == "F"]
girls
```

By using the following figure we can easily find out about the dominant of each gender for each year. As it can be seen, after the male gender was always the dominant of the year.

```python
indices = np.arange(1880, 2015, 1)
width = 0.20
plt.figure(figsize=(25, 12))
plt.bar(indices - width/2, boys["Count"], width, color='IndianRed', label='Boys')
plt.bar(indices + width/2, girls["Count"], width, color='SkyBlue', label='Girls')

plt.xlabel("Year")
plt.ylabel("Number of boys and girls")
plt.legend()
```

In the next step I am going to work on the most trended names in the last 10 years in the USA. At first I use the "groupby" feature to find the recently trended names. The problem is that I can find just the numbers by using this method. To solve this issue I will create a dictionary based on what I found first, then I use this dictionary to find the "top_10" which is the data set I am looking for. Based on the "top_10" I can see that which name is most trended in each year of this period.

```python
recently_trended = national_names[(national_names["Year"] > 2004) & (national_names["Year"] <= 2014)]
f = recently_trended.groupby("Year")["Count"].max()
f = f.reset_index()
f
```

```python
r = list(f["Year"])
w = list(f["Count"])
q = {}
for i in range(len(r)):
    q[r[i]] = w[i]

q
```

```python
top_10 = pd.DataFrame(index=range(0))
```

```python
for key,value in q.items():
    top_10 = top_10.append(recently_trended[(recently_trended["Year"] == key) & (recently_trended["Count"] == value)])
    
top_10
```

```python
top_10.plot.bar(x="Name", y="Count")
```
